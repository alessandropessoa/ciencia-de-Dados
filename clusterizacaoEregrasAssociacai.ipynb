```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.cluster import KMeans
```

```python
dfCompleto_
```

```python
dfAgrupamento = dfCompleto_[['sexo','valor_total_compra']]
```

```python
dfAgrupamento
```

```python
padronizacao = StandardScaler()
```

```python
dfAgrupamento_Padronizado = padronizacao.fit_transform(dfAgrupamento)
```

```python
dfAgrupamento_Padronizado
```

```python
dfAgrupamento_Padronizado.shape
```

```python
cluster = KMeans(n_clusters=4,random_state=0)
```

```python
agrupamentoPorRegistro = cluster.fit_predict(dfAgrupamento_Padronizado)
```

```python
dfCompleto_Agrupado =dfCompleto_

```

```python
dfCompleto_Agrupado['cluster'] =  agrupamentoPorRegistro
```

```python
dfCompleto_Agrupado
```

```python
centroides = cluster.cluster_centers_ #centroide dos clusters ,com vaalores escalonados  
centroides
```

```python
centroides_caloresReais =padronizacao.inverse_transform(cluster.cluster_centers_) #valores reais (despadronizados) 
centroides_caloresReais
```

```python
graf = px.scatter(x=dfCompleto_Agrupado['idade'],y=dfCompleto_Agrupado['valor_total_compra'],color=cluster.labels_)
graff2 = px.scatter(x=centroides_caloresReais[:,0],y=centroides_caloresReais[:,1],size=[8,8,8,8],color=range(4))
grafico = go.Figure(data=graf.data+graff2.data)
grafico.update_layout(title_text="Cluters",)

grafico.show()
```

```python
dsMercado_ = pd.read_csv('../desafioFinal/datasets/mercado.csv',sep=';',header=None)
dsMercado_
```

```python
listaMercado = []
```

```python
import time
inicio = time.time()
for x in range(len(dsMercado_)):
    listaMercado.append([str(dsMercado_.values[x,y]) for y in range(1,21)])
tempo = time.time() - inicio
print('tempo para criar lista ',tempo)
```

```python
len(listaMercado)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_array = te.fit(listaMercado).transform(listaMercado)
```

```python
te_array
```

```python
te.columns_
```

```python
len(te.columns_)
```

```python
dfMercadoPreproc = pd.DataFrame(te_array,columns=te.columns_)
```

```python
dfMercadoPreproc
```

```python
#como cerfificamos que havia dados faltantes eles tbm foram tranformado em colunas portanto seram removidos

dfMercadoPreproc.drop(columns='nan',inplace=True)
```

```python
dfMercadoPreproc
```

```python
# Gerando os itens fequentes do dataset
from mlxtend.frequent_patterns import apriori
```

```python
frequenteItensSet_minSup_10 = apriori(dfMercadoPreproc,min_support=0.05,use_colnames=True) # seleciona apenas suporte de 0.005 ou 0,5%
```

```python
frequenteItensSet_minSup_10
```

```python
frequenteItensSet_minSup_10.sort_values(by='support')
```

```python
# tendo os suportes e os itemsets, vamos criar as regras com uma confiançaque definiremos
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
regras = association_rules(frequenteItensSet_minSup_10, metric='confidence', min_threshold=0.25) #confiança de 20% das vezes de ocorrerem
```
